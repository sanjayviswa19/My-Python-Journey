```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()


data["GRADE"] = label_encoder.fit_transform(data["GRADE"])
data
```

```python
#Mean = the average value (the sum of all values divided by number of values). 
#Median = the value in the middle, after you have sorted all values ascending.
#Mode = the value that appears most frequently.
```

```python
X11=data["English"].mean()
X11
```

```python
data["English"].fillna(X11, inplace = True)
```

```python
x1=data["ENGLISH"].mean()
x1
```

```python
x2=data["MATHS"].mean()
x2
```

```python
x3=data["SOCIAL"].mean()
x3
```

```python
x4=data["TOTAL"].mean()
x4
```

```python
x6=data["GRADE"].mode()
x6
```

```python
data["TAMIL"].fillna(X11,inplace=True)
```

```python

data["chol"].fillna(x2, inplace = True)
data["thalach"].fillna(x3, inplace = True)
data["slope"].fillna(x4, inplace = True)
data["ca"].fillna(x5, inplace = True)
data["target"].fillna(x6, inplace = True)
```
